```python
import csv
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

def main():
    ylist1=[0,0,0,0]
    ylist2=[0,0,0,0]


    f=open('2008gender.csv', 'r', encoding='ANSI')
    data = csv.reader(f)
    header = next(data)
    i=0
    for row in data:
        if i==0:
            for a in range(3,50,4):
                if row[a] != ' ':
                    ylist1[0]+=int(row[a].replace(',', ''))
                    ylist2[0]+=int(row[a+1].replace(',', ''))
                a+=3
        i+=1
    f.close()
    f=open('2013gender.csv', 'r', encoding='ANSI')
    data = csv.reader(f)
    header = next(data)
    i=0
    for row in data:
        if i==0:
            for a in range(3,50,4):
                if row[a] != ' ':
                    ylist1[1]+=int(row[a].replace(',', ''))
                    ylist2[1]+=int(row[a+1].replace(',', ''))
                a+=3
        i+=1
    f.close()
    f=open('2018gender.csv', 'r', encoding='ANSI')
    data = csv.reader(f)
    header = next(data)
    i=0
    for row in data:
        if i==0:
            for a in range(3,50,4):
                if row[a] != ' ':
                    ylist1[2]+=int(row[a].replace(',', ''))
                    ylist2[2]+=int(row[a+1].replace(',', ''))
                a+=3
        i+=1
    f.close()
    f=open('2022gender.csv', 'r', encoding='ANSI')
    data = csv.reader(f)
    header = next(data)
    i=0
    for row in data:
        if i==0:
            for a in range(3,50,4):
                if row[a] != ' ':
                    ylist1[3]+=int(row[a].replace(',', ''))
                    ylist2[3]+=int(row[a+1].replace(',', ''))
                a+=3
        i+=1
    f.close()
    print(ylist1)
    print(ylist2)



    xlist=[2008,2013,2018,2022]


    plt.title('제주특별시 년도별 남녀 인구수')
    plt.xlabel('년도')
    plt.ylabel('인구')

    plt.plot(xlist,ylist1,label='남자',c='yellow', marker='o')
    plt.plot(xlist,ylist2,label='여자', c='red', marker='o')

    plt.xticks(xlist)#x리스트 전체 다 나오게 


    for i in range(len(xlist)):
        height = ylist1[i]
        plt.text(xlist[i], height + 0.25, '%.1f' %height, ha='center', va='bottom', size = 10)
    for i in range(len(xlist)):
        height = ylist2[i]
        plt.text(xlist[i], height + 0.25, '%.1f' %height, ha='center', va='top', size = 10)

    plt.subplots_adjust(hspace = 100)

    plt.legend()
    plt.show()

if __name__ == '__main__':

 

    main()
 
```
