```python
import random
import csv


def main():
    hun =[0,0,0,0,0,0]
    tho = [0,0,0,0,0,0]
    tentho = [0,0,0,0,0,0]
    huntho = [0,0,0,0,0,0]

    for i in range(100):
        r = random.randint(1,6) 
        if(r==1):
            hun[0]+=1
        if(r==2):
            hun[1]+=1
        if(r==3):
            hun[2]+=1
        if(r==4):
            hun[3]+=1
        if(r==5):
            hun[4]+=1
        if(r==6):
            hun[5]+=1
    for i in range(1000):
        r = random.randint(1,6) 
        if(r==1):
            tho[0]+=1
        if(r==2):
            tho[1]+=1
        if(r==3):
            tho[2]+=1
        if(r==4):
            tho[3]+=1
        if(r==5):
            tho[4]+=1
        if(r==6):
            tho[5]+=1
    for i in range(10000):
        r = random.randint(1,6) 
        if(r==1):
            tentho[0]+=1
        if(r==2):
            tentho[1]+=1
        if(r==3):
            tentho[2]+=1
        if(r==4):
            tentho[3]+=1
        if(r==5):
            tentho[4]+=1
        if(r==6):
            tentho[5]+=1
    for i in range(100000):
        r = random.randint(1,6) 
        if(r==1):
            huntho[0]+=1
        if(r==2):
            huntho[1]+=1
        if(r==3):
            huntho[2]+=1
        if(r==4):
            huntho[3]+=1
        if(r==5):
            huntho[4]+=1
        if(r==6):
            huntho[5]+=1

    f = open('dice.csv','w', newline='')
    f.write('빈도, 1, 2, 3, 4, 5, 6')
    f.write("\n100")
    for i in hun: 
        f.write(","+str(i))
    f.write("\n1000")
    for i in tho: 
        f.write(","+str(i))
    f.write("\n10000")
    for i in tentho: 
        f.write(","+str(i))
    f.write("\n100000")
    for i in huntho: 
        f.write(","+str(i))
    f.close()
    print("파일 생성 완료")
 
if __name__ == '__main__':

    main()  
```
