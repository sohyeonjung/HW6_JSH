```python
import csv
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

def main():
    on = []
    off = []
    onoff = []
    onstation = []
    oncount = []
    offstation = []
    offcount = []
    onoffstation = []
    onoffcount = []
    
    f = open('sub03.csv', 'r')
    data = csv.reader(f)
    header = next(data)
    for row in data:
        row = list(map(lambda s: ''.join(s.split()), row))

        a = int(row[9]) + int(row[11])
        b = int(row[10]) + int(row[12])
        c = a + b

        a1 = [row[1] + " " + row[2], a]
        a2 = [row[1] + " " + row[2], b]
        a3 = [row[1] + " " + row[2], c]
        on.append(a1)
        off.append(a2)
        onoff.append(a3)

    on.sort(key=lambda x: -x[1])
    off.sort(key=lambda x: -x[1])
    onoff.sort(key=lambda x: -x[1])

    for i in range(30):
        onstation.append(on[i][0])
        oncount.append(on[i][1])
        offstation.append(off[i][0])
        offcount.append(off[i][1])
        onoffstation.append(onoff[i][0])
        onoffcount.append(onoff[i][1])

    f.close()

    fig, axs = plt.subplots(3, 1, figsize=(10, 20))

    axs[0].bar(onstation, oncount)
    axs[0].set_title("승차")
    axs[0].tick_params(axis='x', rotation=90)

    axs[1].bar(offstation, offcount)
    axs[1].set_title("하차")
    axs[1].tick_params(axis='x', rotation=90)

    axs[2].bar(onoffstation, onoffcount)
    axs[2].set_title("승하차")
    axs[2].tick_params(axis='x', rotation=90)

    plt.show()

if __name__ == "__main__":
    main()
```
